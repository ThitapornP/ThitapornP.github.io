```python
# create SQL statements for language_training_center table
create_table_query = 'CREATE TABLE language_training_center (center_id integer PRIMARY KEY,center_name varchar(50) NOT NULL,center_city varchar(30) NOT NULL);'
insert_query_list = ["INSERT INTO language_training_center VALUES(1001,'NNC Language training center','Newcastle upon Tyne');",
                    "INSERT INTO language_training_center VALUES(1002,'DUR Language training center','Durham');"]

try:
    # Drop the table language_training_center to make sure that there's no duplication of it
    cursor.execute("DROP TABLE IF EXISTS language_training_center")
    # Create the table
    cursor.execute(create_table_query)
    # Insert data
    for insert_query in insert_query_list:
        cursor.execute(insert_query)
    # Commit the changes    
    conn.commit()

    # Execute the SELECT query
    cursor.execute("SELECT * FROM language_training_center")
    # Fetch all the results
    results = cursor.fetchall()
    # Print the result
    for result in results:
        print(result)
# Print the error
except mysql.connector.Error as error:
    print("Failed to create the table language_training_center or insert record into language_training_center table {}".format(error))
```

# Q2: The female employment situation in Northeast region 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Import CSV file
NEstaff_2019_20 = pd.read_csv("NEstaff_2019-20.csv")
NEstaff_2020_21 = pd.read_csv("NEstaff_2020-21.csv")
```

```python
# Check info of NEstaff_2019_20 file
NEstaff_2019_20.info()
```

```python
# Check info of NEstaff_2020_21 file
NEstaff_2020_21.info()
```

```python
# Combine 2 files into one dataframe
frames = [NEstaff_2019_20, NEstaff_2020_21]
NEstaff_combine = pd.concat(frames)
NEstaff_combine
```

```python
# Plot barplot: Staff Number in two Academic Year separated by Sex

# Set the figure size
plt.figure(figsize=(7,4))
# Plot the graph
fig_1=sns.barplot(data=NEstaff_combine, x='Academic Year', y="Number", hue='Sex', errwidth=0, 
              estimator=sum, palette="Spectral")
# Set data labels
for i in fig_1.containers:
    fig_1.bar_label(i,)
# Set plot title
plt.title('Staff Number in two Academic Years separated by Sex')
plt.show()
```

```python
# Combine 2 files into one dataframe but only female
Female_NEstaff_combine=NEstaff_combine[NEstaff_combine.Sex=='Female']
Female_NEstaff_combine
```

```python
# Set dataframe of academic year 19/21 but only female
Female_NEstaff_2019_20=NEstaff_2019_20[NEstaff_2019_20.Sex=='Female']
Female_NEstaff_2019_20
```

```python
# Plot barplot: Female Staff Number of Contract salary by Contract levels in 2019/20

# Set the figure size
plt.figure(figsize=(8,6))
# Plot the graph
fig_3=sns.barplot(data=Female_NEstaff_2019_20, x='Number', y="Contract salary", 
              hue="Contract levels", errwidth=0, estimator=sum, palette="Spectral")
# Set data labels
for i in fig_3.containers:
    fig_3.bar_label(i,)
# Set plot title
plt.title('Female Staff Number of Contract salary by Contract levels in 2019/20')
plt.show()
```

```python
# Set dataframe of academic year 20/21 but only female
Female_NEstaff_2020_21=NEstaff_2020_21[NEstaff_2020_21.Sex=='Female']
Female_NEstaff_2020_21
```

```python
# Plot barplot: Female Staff Number of Contract salary by Contract levels in 2019/20

# Set the figure size
plt.figure(figsize=(8,6))
# Plot the graph
fig_5=sns.barplot(data=Female_NEstaff_2020_21, x='Number', y="Contract salary", 
              hue="Contract levels", errwidth=0, estimator=sum, palette="Spectral")
# Set data labels
for i in fig_5.containers:
    fig_5.bar_label(i,)
# Set plot title
plt.title('Female Staff Number of Contract salary by Contract levels in 2020/21')
plt.show()
```
